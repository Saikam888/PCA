```python
#Dimensionality Reduction with PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
```

```python
# Standardize the data
features = df.select_dtypes(include=[np.number])
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)
```

```python
scaled_data
```

```python
# Apply PCA
pca = PCA()
pca_components = pca.fit_transform(scaled_data)
```

```python
pca_components
```

```python
# Scree plot
plt.figure(figsize=(10, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance by PCA Components")
plt.grid(True)
plt.show()

```

```python
# Choose number of components (e.g., 2 or 3)
n_components = 2
pca_final = PCA(n_components=n_components)
pca_data = pca_final.fit_transform(scaled_data)
```

```python
pca_data
```

```python
#Clustering with Original Data
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
# Apply KMeans
kmeans_orig = KMeans(n_clusters=3, random_state=42)
clusters_orig = kmeans_orig.fit_predict(scaled_data)
```

```python
clusters_orig
```

```python
# Visualization
plt.figure(figsize=(8, 6))
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=clusters_orig, cmap='viridis')
plt.title("KMeans Clustering on Original Data")
plt.show()
```

```python
# Evaluation
silhouette_orig = silhouette_score(scaled_data, clusters_orig)
dbi_orig = davies_bouldin_score(scaled_data, clusters_orig)
print(f"Silhouette Score (Original): {silhouette_orig:.2f}")
print(f"Daviesâ€“Bouldin Index (Original): {dbi_orig:.2f}")
```

```python
#Clustering with PCA Data
# Apply KMeans on PCA data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(pca_data)
```

```python
clusters_pca
```

```python
# Visualization
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=clusters_pca, cmap='plasma')
plt.title("KMeans Clustering on PCA-Transformed Data")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# Evaluation
silhouette_pca = silhouette_score(pca_data, clusters_pca)
dbi_pca = davies_bouldin_score(pca_data, clusters_pca)
print(f"Silhouette Score (PCA): {silhouette_pca:.2f}")
print(f"Daviesâ€“Bouldin Index (PCA): {dbi_pca:.2f}")
```

```python
#Comparison and Analysis
print(f"Silhouette Score Comparison: Original = {silhouette_orig:.2f}, PCA = {silhouette_pca:.2f}")
print(f"Daviesâ€“Bouldin Index Comparison: Original = {dbi_orig:.2f}, PCA = {dbi_pca:.2f}")
```
